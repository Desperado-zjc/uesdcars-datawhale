训练数据集一共有15万行31列，其中price列是价格，其余30列是特征参数。因为30个特种不是很多，我们不妨一个一个特征来分析，以熟悉数据集。<br/>
**SaleID** 与index一致，顺序递增，从0-149999<br/>
**name** 名字？具体是什么名字我们也不知道，可能是买方的名字，或者卖方的名字，或者其他……<br/>
**regDate** 注册日期，为了便于训练，不妨设定一个最早的日期为时间0，后续用日期相关函数进行相减法，或者其他的一些时间处理方法<br/>
**model** 车的型号，也许训练的时候型号要与车的品牌进行关联<br/>
**brand** 车的品牌<br/>
**bodyType** 车型，应该指的是轿车或是面包车之类的<br/>
**fuelType** 加油的型号<br/>
**gearbox** 变速箱<br/>
**power** 车的马力<br/>
**kilometer** 估计是每升油可以跑多少公里<br/>
**notRepairedDamage** 无修理损伤<br/>
**regionCode** 车牌所在地<br/>
**seller** 只有0和1两种状态，不大清楚是什么<br/>
**offerType** 貌似全是0啊<br/>
**creatDate** 应该说的是什么时候放在网上售卖<br/>
**price** 价格情况，最低11最高99999，感觉是被脱敏了<br/>
**v01-v14** 其他的特征值<br/>

```python
train_data.loc[:,'offerType'].value_counts() # 这个函数用于对每一列的数据统计各数值重复出现的次数……果然offerType全是0
```
